```python
import numpy as np
from mathlol import mathlol

import time
import random
import matplotlib.pyplot as plt
```

```python
data_for_mlol = {}
for i in range(10, 100, 10):
    array = []
    for i_ in range(i):
        temp = []
        for j_ in range(i):
            temp.append(random.randint(-100, 100))
        array.append(temp)
    data_for_mlol[i] = array
```

```python
data_for_numpy = {}
for i in range(100, 1000, 100):
    array = []
    for i_ in range(i):
        temp = []
        for j_ in range(i):
            temp.append(random.randint(-100, 100))
        array.append(temp)
    data_for_numpy[i] = array
```

```python
numpy = {}
for key in data_for_numpy.keys():
    start = time.process_time()
    matrix = np.array(data_for_numpy[key])
    result = matrix * matrix
    end = time.process_time() - start
    numpy[key] = end
```

```python
mlol = {}
for key in data_for_mlol.keys():
    start = time.process_time()
    matrix = mathlol()
    matrix.set(matrix = data_for_mlol[key])
    result = matrix * matrix
    end = time.process_time() - start
    mlol[key] = end
```

```python
plt.plot(numpy.values())
plt.plot(mlol.values())

plt.title("NumPy vs MathLOL")
plt.legend(["NumPy (x10)", "MathLOL"])

plt.ylabel("Время (сек)")
plt.xticks([])
```
